```python
from pysus.online_data import SIA
import pandas as pd

# 1. Configurações
ano = 2024
estado = 'TO'
# Municípios da Região Sul (Gurupi e arredores)
municipios_sul = ['170950', '170070', '170820', '171660', '170110']

dfs_meses = []

print(f"Iniciando extração para {estado} {ano} com argumentos corrigidos...")

for mes in range(1, 13):
    try:
        # A versão atual do seu PySUS exige 'groups' no plural
        data = SIA.download(states=estado, years=ano, months=mes, groups='PA')

        df_temp = data.to_dataframe()

        # Filtro de Municípios
        df_temp['PA_UFMUN'] = df_temp['PA_UFMUN'].astype(str)
        filtrado = df_temp[df_temp['PA_UFMUN'].isin(municipios_sul)].copy()

        if not filtrado.empty:
            dfs_meses.append(filtrado)
            print(f"Mês {mes}: {len(filtrado)} registros filtrados.")
        else:
            print(f"Mês {mes}: Arquivo lido, mas sem dados para os municípios da Região Sul.")

    except Exception as e:
        print(f"Mês {mes}: Erro -> {e}")

# 3. Consolidação e Filtro de Exames
if dfs_meses:
    df_final = pd.concat(dfs_meses, ignore_index=True)

    # Procedimentos de Diagnóstico Laboratorial começam com '02'
    df_final['PA_PROC_ID'] = df_final['PA_PROC_ID'].astype(str)
    df_exames = df_final[df_final['PA_PROC_ID'].str.startswith('02')].copy()

    print("\n" + "="*40)
    print(f"RESULTADO PARA REGIAO SUL (2024)")
    print(f"Total de registros de exames: {len(df_exames)}")
    print("="*40)

    # Exibe os códigos dos exames mais comuns
    print("\nTop 5 códigos de exames realizados:")
    print(df_exames['PA_PROC_ID'].value_counts().head())
else:
    print("\nNenhum dado foi processado. Verifique se o PySUS está atualizado.")
```

```python
from pysus.online_data import SIA
import pandas as pd

# 1. Configurações
ano = 2024
estado = 'TO'
municipios_sul = ['170950', '170070', '170820', '171660', '170110']

# 2. Dicionário de nomes (Exemplos comuns do seu modelo)
# No mundo ideal, você cruzaria com a tabela SIGTAP,
# mas aqui faremos o mapeamento para as colunas da sua planilha.
dfs_meses = []

print(f"Extraindo dados de {estado} - {ano}...")

for mes in range(1, 13):
    try:
        # Baixando com a sintaxe que funcionou (plural)
        data = SIA.download(states=estado, years=ano, months=mes, groups='PA')
        df_temp = data.to_dataframe()

        # Filtro de Municípios e Procedimentos Laboratoriais (Grupo 02)
        df_temp['PA_UFMUN'] = df_temp['PA_UFMUN'].astype(str)
        df_temp['PA_PROC_ID'] = df_temp['PA_PROC_ID'].astype(str)

        mask = (df_temp['PA_UFMUN'].isin(municipios_sul)) & (df_temp['PA_PROC_ID'].str.startswith('02'))
        filtrado = df_temp[mask].copy()

        if not filtrado.empty:
            dfs_meses.append(filtrado)
            print(f"Mês {mes}: {len(filtrado)} registros.")

    except Exception as e:
        print(f"Erro no mês {mes}: {e}")

# 3. Formatação estilo "Modelo de Planilha"
if dfs_meses:
    df_raw = pd.concat(dfs_meses, ignore_index=True)

    # Criando a estrutura da sua planilha: "Procedimento", "Qtd" e "Total"
    # Nota: No SIA, PA_QTDPRO é a quantidade aprovada.
    df_export = df_raw.groupby('PA_PROC_ID')['PA_QTDPRO'].sum().reset_index()

    # Renomeando colunas para o seu padrão
    df_export.columns = ['Procedimento', 'Qtd.aprovada']
    df_export['Total'] = df_export['Qtd.aprovada']

    # Salvando no formato CSV com ponto e vírgula (igual ao seu arquivo)
    nome_arquivo = f"SIA_SUL_TO_{ano}_Formatado.csv"
    df_export.to_csv(nome_arquivo, sep=';', index=False, encoding='utf-8-sig')

    print("\n" + "="*30)
    print(f"ARQUIVO GERADO: {nome_arquivo}")
    print(f"Total de linhas no modelo: {len(df_export)}")
    print("="*30)
else:
    print("Nenhum dado encontrado para processar.")
```

```python
from pysus.online_data import SIA
import pandas as pd

# 1. Lista completa dos 18 municípios da Região Sul (Amor Perfeito - Polo Gurupi)
municipios_sul_completo = [
    '170070', # Alvorada
    '170110', # Araguaçu
    '170460', # Cariri do Tocantins
    '170610', # Crixás do Tocantins
    '170740', # Figueirópolis
    '170820', # Formoso do Araguaia
    '170950', # Gurupi (Polo)
    '170980', # Jaú do Tocantins
    '171090', # Jaú do Tocantins (ou similar na região)
    '171660', # Peixe
    '171700', # Pium
    '171820', # Palmeirópolis
    '171888', # Santa Rita do Tocantins
    '171889', # Sandolândia
    '171890', # Santa Rosa do Tocantins
    '171900', # São Salvador do Tocantins
    '172020', # São Valério
    '172049'  # Sucupira
]

ano = 2024
estado = 'TO'
dfs_meses = []

print(f"Iniciando extração dos 18 municípios do Sul de TO - Ano {ano}")

# 2. Loop de Download (Sintaxe Plural que funcionou no seu ambiente)
for mes in range(1, 13):
    try:
        print(f"Baixando mês {mes}...")
        data = SIA.download(states=estado, years=ano, months=mes, groups='PA')
        df_temp = data.to_dataframe()

        # Filtro: Municípios da lista E Procedimentos que começam com '02'
        df_temp['PA_UFMUN'] = df_temp['PA_UFMUN'].astype(str)
        df_temp['PA_PROC_ID'] = df_temp['PA_PROC_ID'].astype(str)

        filtrado = df_temp[
            (df_temp['PA_UFMUN'].isin(municipios_sul_completo)) &
            (df_temp['PA_PROC_ID'].str.startswith('02'))
        ].copy()

        if not filtrado.empty:
            dfs_meses.append(filtrado)

    except Exception as e:
        print(f"Mês {mes}: Erro -> {e}")

# 3. Transformação para o Formato da sua Planilha
if dfs_meses:
    df_consolidado = pd.concat(dfs_meses, ignore_index=True)

    # Agrupar para somar as quantidades (como no seu modelo)
    # A coluna PA_QTDPRO é a "Qtd.aprovada"
    df_modelo = df_consolidado.groupby('PA_PROC_ID')['PA_QTDPRO'].sum().reset_index()

    # Renomear para o formato da sua planilha
    df_modelo.columns = ['Procedimento', 'Qtd.aprovada']
    df_modelo['Total'] = df_modelo['Qtd.aprovada']

    # Salvar em CSV (usando ponto e vírgula e aspas para texto, igual ao original)
    nome_saida = f"datasus_SUL_TO_2024.csv"
    df_modelo.to_csv(nome_saida, sep=';', index=False, encoding='utf-8-sig', quoting=1)

    print("\n" + "="*40)
    print(f"SUCESSO!")
    print(f"Arquivo gerado: {nome_saida}")
    print(f"Total de tipos de exames realizados: {len(df_modelo)}")
    print(f"Soma total de exames aprovados: {df_modelo['Total'].sum()}")
    print("="*40)
else:
    print("Falha: Nenhum dado foi processado.")
```
