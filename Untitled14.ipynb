```python
from pysus.online_data import SIH
import pandas as pd

# 1. Baixar os dados
# O parâmetro 'groups' é obrigatório. Para o SIH-RD, usamos 'RD'.
# 'states' e 'years' agora aceitam listas ou valores únicos.
dataset = SIH.download(states='MG', years=2024, months=1, groups='RD')

# O PySUS retorna um objeto que contém os arquivos baixados.
# Para transformar em um DataFrame do Pandas:
df = dataset.to_dataframe()

# 2. Filtrar pelo CID-10 N39.0
# Importante: No DATASUS, o ponto costuma ser omitido (N390)
itu_mg = df[df['DIAG_PRINC'] == 'N390']

print(f"Total de internações por ITU em MG (Jan/2024): {len(itu_mg)}")
```

```python
from pysus.online_data import SIH
import pandas as pd

def extrair_contagem_itu(uf, ano, meses):
    resultados = []
    for mes in meses:
        print(f"Processando {uf} - {mes}/{ano}...")
        try:
            # Baixa os dados do mês específico
            dataset = SIH.download(states=uf, years=ano, months=mes, groups='RD')
            df = dataset.to_dataframe()

            # Filtra por ITU (N39.0)
            contagem = len(df[df['DIAG_PRINC'] == 'N390'])

            resultados.append({'Mes': mes, 'UF': uf, 'Internacoes': contagem})
        except Exception as e:
            print(f"Erro no mês {mes}: {e}")

    return pd.DataFrame(resultados)

# Meses de Janeiro a Dezembro
meses_2024 = list(range(1, 13))

# Extrair para MG e AC
df_mg = extrair_contagem_itu('MG', 2024, meses_2024)
df_ac = extrair_contagem_itu('AC', 2024, meses_2024)

# Unir os dados para facilitar a plotagem
df_final = pd.concat([df_mg, df_ac])
print(df_final)
```
