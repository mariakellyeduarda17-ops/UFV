```python
import pandas as pd
import os
from tqdm import tqdm
from pysus.online_data.SIH import SIH

def extrair_dados_itu():
    ufs = ['AC', 'AL', 'AP', 'AM', 'BA', 'CE', 'DF', 'ES', 'GO', 'MA',
           'MT', 'MS', 'MG', 'PA', 'PB', 'PR', 'PE', 'PI', 'RJ', 'RN',
           'RS', 'RO', 'RR', 'SC', 'SP', 'SE', 'TO']

    if not os.path.exists('dados_ITU'):
        os.makedirs('dados_ITU')

    sih = SIH()

    for ano in range(2015, 2025):
        print(f"\n>>> Processando ano: {ano}")
        base_anual = []

        for uf in tqdm(ufs, desc="Progresso"):
            try:
                # Download dos dados
                dataset = sih.download(states=uf, years=ano, months=list(range(1, 13)), group='RD')
                df = dataset.to_dataframe()

                # Filtro por ITU (CID N390)
                mask = df['DIAG_PRINC'].str.contains('N390', na=False)
                df_itu = df[mask].copy()

                if not df_itu.empty:
                    base_anual.append(df_itu)
            except:
                continue

        # Salva o ano processado
        if base_anual:
            df_final_ano = pd.concat(base_anual, ignore_index=True)
            nome_arq = f"dados_ITU/ITU_Brasil_{ano}.csv"
            df_final_ano.to_csv(nome_arq, index=False, encoding='utf-8-sig')
            total = len(df_final_ano)
            print(f"Concluido: {nome_arq} | Total: {total}")

    print("\n--- PROCESSO FINALIZADO ---")

extrair_dados_itu()
```

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from tqdm import tqdm
from pysus.online_data.SIH import SIH
import glob

# 1. Configurações Iniciais
if not os.path.exists('dados_ITU'):
    os.makedirs('dados_ITU')

sih = SIH()
# Lista reduzida com estados que representam bem o Brasil para evitar queda de conexão
# Se quiser todos, use a lista completa anterior
ufs_principais = ['SP', 'RJ', 'MG', 'RS', 'PR', 'BA', 'SC', 'CE', 'PE', 'GO']

# 2. Loop de Extração (2015-2024)
print("--- INICIANDO EXTRAÇÃO ---")
for ano in range(2015, 2025):
    arquivo_final = f"dados_ITU/ITU_{ano}.csv"

    # Só extrai se o arquivo ainda não existir (economiza tempo se você rodar de novo)
    if not os.path.exists(arquivo_final):
        base_ano = []
        for uf in tqdm(ufs_principais, desc=f"Extraindo {ano}"):
            try:
                # Baixando dados (RD = Internações)
                dataset = sih.download(states=uf, years=ano, months=list(range(1, 13)), group='RD')
                df_temp = dataset.to_dataframe()

                # Filtrando CID N39.0 (Infecção Urinária)
                df_itu = df_temp[df_temp['DIAG_PRINC'].str.contains('N390', na=False)].copy()

                if not df_itu.empty:
                    # Guardamos apenas o necessário para o gráfico para não estourar a RAM
                    base_ano.append(df_itu[['ANO_CMPT', 'MES_CMPT', 'DIAG_PRINC']])
            except:
                continue

        if base_ano:
            pd.concat(base_ano, ignore_index=True).to_csv(arquivo_final, index=False)
            print(f"✅ Ano {ano} salvo.")
    else:
        print(f"⏩ Ano {ano} já existe, pulando...")

# 3. Gerar o Gráfico
print("\n--- GERANDO VISUALIZAÇÃO ---")
arquivos = glob.glob("dados_ITU/ITU_*.csv")

if not arquivos:
    print("❌ Erro: Nenhum arquivo foi gerado. Verifique sua conexão com a internet.")
else:
    # Ler e tratar
    df_lista = [pd.read_csv(f) for f in arquivos]
    df_final = pd.concat(df_lista, ignore_index=True)

    # Criar coluna de Data
    df_final['Data'] = pd.to_datetime(
        df_final['ANO_CMPT'].astype(str) + '-' + df_final['MES_CMPT'].astype(str) + '-01'
    )

    # Agrupar por mês
    serie = df_final.groupby('Data').size().reset_index(name='Casos').sort_values('Data')

    # Plotagem
    plt.figure(figsize=(14, 6))
    sns.set_style("whitegrid")

    # Linha principal
    plt.plot(serie['Data'], serie['Casos'], marker='o', markersize=3, label='Internações Reais', alpha=0.6)

    # Média móvel para suavizar e ver a tendência (janela de 12 meses)
    if len(serie) >= 12:
        plt.plot(serie['Data'], serie['Casos'].rolling(12).mean(), color='red', linewidth=3, label='Tendência (Média 12 Meses)')

    plt.title('Internações por Infecção Urinária (ITU) no Brasil | 2015-2024', fontsize=14)
    plt.ylabel('Número de Internações')
    plt.xlabel('Ano')
    plt.legend()
    plt.tight_layout()
    plt.show()
```
