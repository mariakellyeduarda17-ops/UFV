```python
from pysus.online_data import SIA
import pandas as pd

# 1. Configurações básicas
ano = 2024
estado = 'TO'
# Principais cidades da Região Sul (Gurupi, Alvorada, Formoso do Araguaia)
municipios_sul = ['170950', '170070', '170820', '171660', '170110']

dfs_meses = []

print(f"Iniciando busca para {estado} em {ano}...")

# 2. Loop de download
for mes in range(1, 13):
    try:
        # Download dos dados
        data = SIA.download(states=estado, years=ano, months=mes, group='PA')
        df_temp = data.to_dataframe()

        # Filtro imediato por município (economiza memória)
        df_temp['PA_UFMUN'] = df_temp['PA_UFMUN'].astype(str)
        filtrado = df_temp[df_temp['PA_UFMUN'].isin(municipios_sul)].copy()

        if not filtrado.empty:
            dfs_meses.append(filtrado)
            print(f"Mês {mes}: {len(filtrado)} registros encontrados.")
        else:
            print(f"Mês {mes}: Nenhum dado para a região sul.")

    except Exception as e:
        print(f"Mês {mes}: Erro no processamento - {e}")

# 3. Consolidação final
if dfs_meses:
    df_final = pd.concat(dfs_meses, ignore_index=True)

    # Filtro para exames laboratoriais (Grupo 02 do SUS)
    df_final['PA_PROC_ID'] = df_final['PA_PROC_ID'].astype(str)
    df_exames = df_final[df_final['PA_PROC_ID'].str.startswith('02')].copy()

    print("\n" + "="*40)
    print(f"RESULTADO FINAL")
    print(f"Total de exames na Região Sul em 2024: {len(df_exames)}")
    print("="*40)

    # Exibe os 10 exames mais realizados
    print("\nTop 10 procedimentos (Códigos):")
    print(df_exames['PA_PROC_ID'].value_counts().head(10))
else:
    print("\nNenhum dado foi baixado com sucesso.")
```

```python
from pysus.online_data import SIA
import pandas as pd

# 1. Configurações
ano = 2024
estado = 'TO'
municipios_sul = ['170950', '170070', '170820', '171660', '170110']

dfs_meses = []

print(f"Iniciando extração simplificada para {estado} {ano}...")

for mes in range(1, 13):
    try:
        # REMOVIDO o argumento 'group' que causou o erro anterior
        # Mantendo os plurais que seu sistema reconheceu
        data = SIA.download(states=estado, years=ano, months=mes)

        df_temp = data.to_dataframe()

        # Filtro de Municípios
        df_temp['PA_UFMUN'] = df_temp['PA_UFMUN'].astype(str)
        filtrado = df_temp[df_temp['PA_UFMUN'].isin(municipios_sul)].copy()

        if not filtrado.empty:
            dfs_meses.append(filtrado)
            print(f"Mês {mes}: {len(filtrado)} linhas encontradas.")
        else:
            print(f"Mês {mes}: Sem dados para a Região Sul.")

    except Exception as e:
        print(f"Mês {mes}: Falhou com erro -> {e}")

# 3. Consolidação
if dfs_meses:
    df_final = pd.concat(dfs_meses, ignore_index=True)

    # Filtro: Somente exames (Códigos que iniciam com '02')
    df_final['PA_PROC_ID'] = df_final['PA_PROC_ID'].astype(str)
    df_exames = df_final[df_final['PA_PROC_ID'].str.startswith('02')].copy()

    print("\n" + "="*30)
    print(f"TOTAL DE EXAMES: {len(df_exames)}")
    print("="*30)
    print("\nExemplos de exames encontrados:")
    print(df_exames[['PA_CMP', 'PA_PROC_ID', 'PA_QTDPRO']].head())
else:
    print("\nO servidor do DATASUS não retornou dados. Verifique sua conexão.")
```

```python
from pysus.online_data import SIA
import pandas as pd

# 1. Configurações
ano = 2024
estado = 'TO'
# Municípios da Região Sul (Gurupi e arredores)
municipios_sul = ['170950', '170070', '170820', '171660', '170110']

dfs_meses = []

print(f"Iniciando extração para {estado} {ano} com argumentos corrigidos...")

for mes in range(1, 13):
    try:
        # A versão atual do seu PySUS exige 'groups' no plural
        data = SIA.download(states=estado, years=ano, months=mes, groups='PA')

        df_temp = data.to_dataframe()

        # Filtro de Municípios
        df_temp['PA_UFMUN'] = df_temp['PA_UFMUN'].astype(str)
        filtrado = df_temp[df_temp['PA_UFMUN'].isin(municipios_sul)].copy()

        if not filtrado.empty:
            dfs_meses.append(filtrado)
            print(f"Mês {mes}: {len(filtrado)} registros filtrados.")
        else:
            print(f"Mês {mes}: Arquivo lido, mas sem dados para os municípios da Região Sul.")

    except Exception as e:
        print(f"Mês {mes}: Erro -> {e}")

# 3. Consolidação e Filtro de Exames
if dfs_meses:
    df_final = pd.concat(dfs_meses, ignore_index=True)

    # Procedimentos de Diagnóstico Laboratorial começam com '02'
    df_final['PA_PROC_ID'] = df_final['PA_PROC_ID'].astype(str)
    df_exames = df_final[df_final['PA_PROC_ID'].str.startswith('02')].copy()

    print("\n" + "="*40)
    print(f"RESULTADO PARA REGIAO SUL (2024)")
    print(f"Total de registros de exames: {len(df_exames)}")
    print("="*40)

    # Exibe os códigos dos exames mais comuns
    print("\nTop 5 códigos de exames realizados:")
    print(df_exames['PA_PROC_ID'].value_counts().head())
else:
    print("\nNenhum dado foi processado. Verifique se o PySUS está atualizado.")
```
